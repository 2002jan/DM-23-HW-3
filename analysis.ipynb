## Analysis

```python
import seaborn as sns
import pandas as pd
```

```python
clustered_data = pd.read_csv("data/data_norm_hot_svd_kmeans.csv")

clustered_data.head()
```

```python
sns.displot(clustered_data, x="cluster")
```

```python
for i in range(1, 7):
    sns.displot(clustered_data, x=f"pc_{i}", hue="cluster", kind="kde", fill=True)
```

```python
sns.scatterplot(clustered_data, x="pc_3", y="pc_5", hue="cluster")
```

```python
heatmap_data = clustered_data.drop('id', axis=1)
heatmap_data = heatmap_data.groupby('cluster').mean()

sns.heatmap(heatmap_data)
```
